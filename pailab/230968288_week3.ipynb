Experiment 1: Smart Thermostat Agent (Reactive Agent)

```python
import random

class SmartThermostatAgent:
    def __init__(self):
        self.states = ["Low", "Comfortable", "High"]
        self.temperature = random.choice(self.states)

    def perceive(self):
        return self.temperature

    def act(self):
        if self.temperature == "Low":
            return "Heat"
        if self.temperature == "High":
            return "Cool"
        return random.choice(["Idle", "Heat", "Cool"])

    def update(self, action):
        i = self.states.index(self.temperature)
        if action == "Heat" and i < 2:
            self.temperature = self.states[i + 1]
        elif action == "Cool" and i > 0:
            self.temperature = self.states[i - 1]

    def simulate(self, steps):
        log = []
        for _ in range(steps):
            state = self.perceive()
            action = self.act()
            self.update(action)
            log.append((state, action, self.temperature))
        return log


SmartThermostatAgent().simulate(10)
```

Experiment 2: Elevator Control Agent (Goal-Based Agent)

```python
import random

class ElevatorAgent:
    def __init__(self):
        self.current_floor = random.randint(1, 3)
        self.goal_floor = None

    def perceive_request(self):
        self.goal_floor = random.randint(1, 3)

    def act(self):
        if self.current_floor < self.goal_floor:
            return "Move Up"
        if self.current_floor > self.goal_floor:
            return "Move Down"
        return "Stay"

    def update(self, action):
        if action == "Move Up":
            self.current_floor += 1
        elif action == "Move Down":
            self.current_floor -= 1

    def simulate(self, steps):
        history = []
        for _ in range(steps):
            self.perceive_request()
            while self.current_floor != self.goal_floor:
                action = self.act()
                self.update(action)
                history.append((self.current_floor, action, self.goal_floor))
            history.append((self.current_floor, "Arrived", self.goal_floor))
        return history


ElevatorAgent().simulate(5)
```

Experiment 3: Autonomous Traffic Signal Agent

```python
import random
class TrafficSignalAgent:
    def __init__(self):
        self.utility = {
            ("High", "Low"): 10,
            ("Low", "High"): 10,
            ("High", "High"): 5,
            ("Low", "Low"): 1
        }

    def perceive(self):
        return random.choice(["Low", "High"]), random.choice(["Low", "High"])

    def decide(self, perception):
        a, b = perception
        if a == "High" and b == "Low":
            return "Green A"
        if b == "High" and a == "Low":
            return "Green B"
        return random.choice(["Green A", "Green B"])

    def simulate(self, steps):
        results = []
        for _ in range(steps):
            perception = self.perceive()
            decision = self.decide(perception)
            utility = self.utility[perception]
            results.append((perception, decision, utility))
        return results


TrafficSignalAgent().simulate(10)
```

Experiment 4: Warehouse Robot Navigation Agent

```python
import random

class WarehouseRobotAgent:
    def __init__(self, size):
        self.grid = {(x, y) for x in range(size) for y in range(size)}
        self.obstacles = set(random.sample(list(self.grid), size))
        self.free = self.grid - self.obstacles
        self.position = random.choice(list(self.free))
        self.target = random.choice(list(self.free))
        self.visited = set()

    def perceive(self):
        return self.position

    def move(self):
        x, y = self.position
        self.visited.add(self.position)
        moves = [(x+1,y),(x-1,y),(x,y+1),(x,y-1)]
        random.shuffle(moves)
        for m in moves:
            if m in self.free and m not in self.visited:
                self.position = m
                return m
        return self.position

    def simulate(self, steps):
        path = []
        for _ in range(steps):
            if self.position == self.target:
                break
            path.append(self.position)
            self.move()
        path.append(self.position)
        return path, self.target


WarehouseRobotAgent(5).simulate(20)
```

Experiment 5: Online Recommendation Agent (Learning Agent)

```python
import random

class RecommendationAgent:
    def __init__(self):
        self.performance = {"A": 0, "B": 0}

    def recommend(self):
        if random.random() < 0.3:
            return random.choice(["A", "B"])
        return max(self.performance, key=self.performance.get)

    def learn(self, product, feedback):
        self.performance[product] += feedback

    def simulate(self, sessions):
        history = []
        for _ in range(sessions):
            product = self.recommend()
            feedback = random.choice([1, -1])
            self.learn(product, feedback)
            history.append((product, feedback, dict(self.performance)))
        return history


RecommendationAgent().simulate(15)
```
