# BFS

```python
from collections import deque
def bfs(graph,start,end):
    queue = deque([[start]])
    visited = set()
    nodes_expan = 0

    while queue:
        path = queue.popleft()
        node = path[-1]

        if node == end:
            return path,nodes_expan
        
        if node not in visited:
            visited.add(node)
            nodes_expan+=1

            for neighbours in graph.get(node,[]):
                new_path = list(path)
                new_path.append(neighbours)
                queue.append(new_path)
    return None,nodes_expan
```

```python
maze = {
    'A': ['B', 'C'],
    'B': ['A', 'D', 'E'],
    'C': ['A', 'F'],
    'D': ['B'],
    'E': ['B', 'F'],
    'F': ['C', 'E', 'G'],
    'G': []
}
start = 'A'
end = 'G'
path, nodes = bfs(maze,start,end)
print(path,nodes)
```

```python
def dfs(graph,start,end,path):
    if start == end:
        return path
    
    for neighbor in graph.get(start,[]):
        if neighbor not in path:
            result = dfs(graph,neighbor,end,path+[neighbor])
            if result is not None:
                return result
            
    return None
```

```python
graph = {
    'A': ['B', 'C'],
    'B': ['D', 'E'],
    'C': ['F'],
    'D': [],
    'E': ['F'],
    'F': ['G'],
    'G': []
}

start = 'A'
goal = 'G'

path = dfs(graph, start, goal, [start])

print("DFS Path:", path)
```

```python
import heapq

# Graph with costs (node: (neighbor, distance))
graph = {'A': [('B', 1), ('C', 4)], 'B': [('C', 2), ('D', 5)], 'C': [('D', 1)], 'D': []}
# Heuristics (our "compass" guess to the goal D)
heuristics = {'A': 7, 'B': 6, 'C': 2, 'D': 0}

def a_star_search(start, goal):
    # Priority queue stores: (total_estimated_cost, current_cost, current_node, path)
    queue = [(heuristics[start], 0, start, [start])]
    
    while queue:
        est_cost, cost, node, path = heapq.heappop(queue)
        
        if node == goal:
            return path
            
        for neighbor, weight in graph.get(node, []):
            new_cost = cost + weight # this calc g(n) node n -> next node path
            # f(n) = g(n) (actual cost) + h(n) (heuristic guess)
            f_n = new_cost + heuristics[neighbor]
            heapq.heappush(queue, (f_n, new_cost, neighbor, path + [neighbor]))

print("A* Path A to D:", a_star_search('A', 'D'))
```
