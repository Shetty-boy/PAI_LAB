```python
class Graph:
    def __init__(self, vertices, edges):
        self.vertices = vertices
        self.edges = edges

    def adjacency_list(self):
        graph = {}
        for v in self.vertices:
            graph[v] = []
        for u, v, w in self.edges:
            graph[u].append((v, w))
        return graph

    def adjacency_matrix(self):
        n = len(self.vertices)
        matrix = [[0 for _ in range(n)] for _ in range(n)]
        for u, v, w in self.edges:
            i = self.vertices.index(u)
            j = self.vertices.index(v)
            matrix[i][j] = w
        return matrix


def define_vertices():
    return [1, 2, 3, 4]


def define_edges():
    return [
        (1, 2, 1),
        (1, 3, 1),
        (2, 3, 3),
        (3, 4, 4),
        (4, 1, 5)
    ]


def print_adjacency_list(adj_list):
    for k in adj_list:
        print(k, ":", adj_list[k])


def print_adjacency_matrix(matrix):
    for row in matrix:
        print(row)


vertices = define_vertices()
edges = define_edges()

g = Graph(vertices, edges)

adj_list = g.adjacency_list()
print_adjacency_list(adj_list)

adj_matrix = g.adjacency_matrix()
print_adjacency_matrix(adj_matrix)
```

```python
import math

class DistanceCalculator:
    def __init__(self, x, y):
        self.x = x
        self.y = y

    def euclidean_distances(self):
        distances = []
        for i in range(len(self.x)):
            distances.append(math.sqrt((self.y[i] - self.x[i]) ** 2))
        return distances


class BubbleSort:
    def sort(self, arr):
        n = len(arr)
        for i in range(n):
            for j in range(0, n - i - 1):
                if arr[j] > arr[j + 1]:
                    arr[j], arr[j + 1] = arr[j + 1], arr[j]
        return arr


def create_list_x():
    return [2, 4, 6, 8]


def create_list_y():
    return [1, 3, 5, 10]


X = create_list_x()
Y = create_list_y()

dc = DistanceCalculator(X, Y)
distances = dc.euclidean_distances()

bs = BubbleSort()
sorted_distances = bs.sort(distances)

print("Distances:", distances)
print("Sorted Distances:", sorted_distances)
```

```python
class DirectedGraph:
    def __init__(self):
        self.graph = {}

    def add_vertex(self, v):
        if v not in self.graph:
            self.graph[v] = []

    def add_edge(self, u, v):
        if u not in self.graph:
            self.graph[u] = []
        if v not in self.graph:
            self.graph[v] = []
        self.graph[u].append(v)

    def adjacency_list(self):
        return self.graph

    def adjacency_matrix(self):
        vertices = list(self.graph.keys())
        n = len(vertices)
        matrix = [[0 for _ in range(n)] for _ in range(n)]
        for i in range(n):
            for j in self.graph[vertices[i]]:
                matrix[i][vertices.index(j)] = 1
        return matrix


def define_vertices():
    return [1, 2, 3, 4]


def define_edges():
    return [(1, 2), (1, 3), (2, 4), (3, 4)]


g = DirectedGraph()

vertices = define_vertices()
edges = define_edges()

for v in vertices:
    g.add_vertex(v)

for u, v in edges:
    g.add_edge(u, v)

adj_list = g.adjacency_list()
adj_matrix = g.adjacency_matrix()

print("Adjacency List")
for k in adj_list:
    print(k, ":", adj_list[k])

print("Adjacency Matrix")
for row in adj_matrix:
    print(row)
```
