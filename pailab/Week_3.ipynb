```python
class Environment:
    def __init__(self, initial_temperature):
        self.temperature = initial_temperature

    def apply_action(self, action):
        if action == "Heat" and self.temperature == "Low":
            self.temperature = "Comfortable"
        elif action == "Heat" and self.temperature == "Comfortable":
            self.temperature = "High"
        elif action == "Cool" and self.temperature == "High":
            self.temperature = "Comfortable"
        elif action == "Cool" and self.temperature == "Comfortable":
            self.temperature = "Low"

    def get_percept(self):
        return self.temperature


class SmartThermostatAgent:
    def __init__(self):
        self.rules = {
            "Low": "Heat",
            "Comfortable": "Idle",
            "High": "Cool"
        }

    def act(self, percept):
        return self.rules[percept]


def simulate(initial_temperature, steps):
    env = Environment(initial_temperature)
    agent = SmartThermostatAgent()
    history = []

    for _ in range(steps):
        percept = env.get_percept()
        action = agent.act(percept)
        history.append((percept, action))
        env.apply_action(action)

    return history


if __name__ == "__main__":
    for temp in ["Low", "Comfortable", "High"]:
        result = simulate(temp, 5)
        print(f"Initial Temperature: {temp}")
        for state, action in result:
            print(state, "->", action)
        print()
```

```python
import random

class ElevatorEnvironment:
    def __init__(self, start_floor=1):
        self.current_floor = start_floor
        self.request_floor = None

    def set_request(self, floor):
        self.request_floor = floor

    def get_percept(self):
        return self.current_floor, self.request_floor

    def apply_action(self, action):
        if action == "Move Up" and self.current_floor < 3:
            self.current_floor += 1
        elif action == "Move Down" and self.current_floor > 1:
            self.current_floor -= 1
        if self.current_floor == self.request_floor:
            self.request_floor = None


class ElevatorAgent:
    def __init__(self, goal_floor=None):
        self.goal_floor = goal_floor

    def formulate_goal(self, request_floor):
        self.goal_floor = request_floor

    def act(self, current_floor):
        if self.goal_floor is None:
            return "Stay"
        if current_floor < self.goal_floor:
            return "Move Up"
        if current_floor > self.goal_floor:
            return "Move Down"
        return "Stay"


def simulate(start_floor, requests, steps):
    env = ElevatorEnvironment(start_floor)
    agent = ElevatorAgent()
    history = []

    for t in range(steps):
        if t < len(requests) and requests[t] is not None:
            env.set_request(requests[t])
            agent.formulate_goal(requests[t])

        current_floor, request_floor = env.get_percept()
        action = agent.act(current_floor)
        history.append((t, current_floor, request_floor, action))
        env.apply_action(action)

    return history


def random_test_cases(num_cases=3, steps=6):
    for i in range(num_cases):
        start_floor = random.randint(1, 3)
        requests = [random.choice([1, 2, 3, None]) for _ in range(steps)]
        result = simulate(start_floor, requests, steps)

        print(f"Test Case {i + 1}")
        print("Start Floor:", start_floor)
        print("Requests:", requests)
        for entry in result:
            print(entry)
        print()


if __name__ == "__main__":
    random_test_cases()
```

```python
import random

class TrafficEnvironment:
    def __init__(self):
        self.road_a = random.choice(["Low", "High"])
        self.road_b = random.choice(["Low", "High"])

    def get_percept(self):
        return self.road_a, self.road_b

    def update_traffic(self):
        self.road_a = random.choice(["Low", "High"])
        self.road_b = random.choice(["Low", "High"])


class TrafficSignalAgent:
    def __init__(self):
        self.utility_map = {
            "Low": 1,
            "High": 5
        }

    def calculate_utility(self, green_road, traffic_a, traffic_b):
        if green_road == "A":
            return -self.utility_map[traffic_b]
        else:
            return -self.utility_map[traffic_a]

    def act(self, traffic_a, traffic_b):
        utility_a = self.calculate_utility("A", traffic_a, traffic_b)
        utility_b = self.calculate_utility("B", traffic_a, traffic_b)
        if utility_a > utility_b:
            return "Green for Road A"
        return "Green for Road B"


def simulate(steps):
    env = TrafficEnvironment()
    agent = TrafficSignalAgent()
    history = []

    for t in range(steps):
        traffic_a, traffic_b = env.get_percept()
        action = agent.act(traffic_a, traffic_b)
        history.append((t, traffic_a, traffic_b, action))
        env.update_traffic()

    return history


def random_test_cases(cases=3, steps=6):
    for i in range(cases):
        print(f"Test Case {i + 1}")
        result = simulate(steps)
        for entry in result:
            print(entry)
        print()


if __name__ == "__main__":
    random_test_cases()
```
