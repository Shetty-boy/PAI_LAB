```python
from collections import deque
import random
import string

class SearchProblem:
    def __init__(self, graph, start, goal):
        self.graph = graph
        self.start = start
        self.goal = goal

    def initial_state(self):
        return self.start

    def goal_test(self, state):
        return state == self.goal

    def actions(self, state):
        return self.graph.get(state, [])

    def transition(self, state, action):
        return action

    def path_cost(self, cost):
        return cost + 1


class BreadthFirstSearch:
    def __init__(self, problem):
        self.problem = problem
        self.nodes_expanded = 0

    def search(self):
        queue = deque([[self.problem.initial_state()]])
        visited = set()

        while queue:
            path = queue.popleft()
            current = path[-1]

            if self.problem.goal_test(current):
                return path, self.nodes_expanded

            if current not in visited:
                visited.add(current)
                self.nodes_expanded += 1

                for action in self.problem.actions(current):
                    new_state = self.problem.transition(current, action)
                    queue.append(path + [new_state])

        return None, self.nodes_expanded


def generate_random_maze(nodes, max_edges=3):
    graph = {}
    for node in nodes:
        connections = random.sample(nodes, random.randint(1, min(max_edges, len(nodes))))
        connections = [c for c in connections if c != node]
        graph[node] = connections
    return graph


def run_random_tests(test_cases=3):
    for i in range(test_cases):
        nodes = random.sample(string.ascii_uppercase, random.randint(5, 8))
        graph = generate_random_maze(nodes)
        start = random.choice(nodes)
        goal = random.choice(nodes)

        problem = SearchProblem(graph, start, goal)
        bfs = BreadthFirstSearch(problem)
        path, expanded = bfs.search()

        print(f"\nTest Case {i + 1}")
        print("Graph:", graph)
        print("Start:", start, "Goal:", goal)
        print("Path:", path)
        print("Nodes Expanded:", expanded)


if __name__ == "__main__":
    maze = {
        'A': ['B', 'C'],
        'B': ['A', 'D', 'E'],
        'C': ['A', 'F'],
        'D': ['B'],
        'E': ['B', 'F'],
        'F': ['C', 'E', 'G'],
        'G': []
    }

    problem = SearchProblem(maze, 'A', 'G')
    bfs = BreadthFirstSearch(problem)
    path, expanded = bfs.search()

    print("Fixed Maze Test")
    print("Path:", path)
    print("Nodes Expanded:", expanded)

    run_random_tests()
```

```python
from collections import deque
import random

class EightPuzzleProblem:
    def __init__(self, initial, goal):
        self.initial = initial
        self.goal = goal

    def initial_state(self):
        return self.initial

    def goal_test(self, state):
        return state == self.goal

    def actions(self, state):
        index = state.index(0)
        row, col = divmod(index, 3)
        moves = []
        if row > 0:
            moves.append(-3)
        if row < 2:
            moves.append(3)
        if col > 0:
            moves.append(-1)
        if col < 2:
            moves.append(1)
        return moves

    def transition(self, state, action):
        new_state = list(state)
        index = state.index(0)
        swap_index = index + action
        new_state[index], new_state[swap_index] = new_state[swap_index], new_state[index]
        return tuple(new_state)


class BreadthFirstSearch:
    def __init__(self, problem):
        self.problem = problem
        self.expanded = 0

    def search(self):
        queue = deque([[self.problem.initial_state()]])
        visited = set([self.problem.initial_state()])

        while queue:
            path = queue.popleft()
            state = path[-1]

            if self.problem.goal_test(state):
                return path, self.expanded

            self.expanded += 1

            for action in self.problem.actions(state):
                next_state = self.problem.transition(state, action)
                if next_state not in visited:
                    visited.add(next_state)
                    queue.append(path + [next_state])

        return None, self.expanded


def generate_solvable_puzzle(moves=20):
    goal = (1, 2, 3, 4, 5, 6, 7, 8, 0)
    state = goal
    problem = EightPuzzleProblem(state, goal)

    for _ in range(moves):
        action = random.choice(problem.actions(state))
        state = problem.transition(state, action)

    return state


def print_solution(path):
    for state in path:
        for i in range(0, 9, 3):
            print(state[i:i+3])
        print()


def run_random_tests(test_cases=2):
    goal = (1, 2, 3, 4, 5, 6, 7, 8, 0)

    for i in range(test_cases):
        initial = generate_solvable_puzzle()
        problem = EightPuzzleProblem(initial, goal)
        bfs = BreadthFirstSearch(problem)
        path, expanded = bfs.search()

        print(f"\nTest Case {i + 1}")
        print("Initial State:")
        print_solution([initial])
        print("Solution Path Length:", len(path) - 1)
        print("States Expanded:", expanded)


if __name__ == "__main__":
    initial_state = (1, 2, 3, 4, 0, 6, 7, 5, 8)
    goal_state = (1, 2, 3, 4, 5, 6, 7, 8, 0)

    problem = EightPuzzleProblem(initial_state, goal_state)
    bfs = BreadthFirstSearch(problem)
    solution, expanded = bfs.search()

    print("Fixed Test Case Solution:")
    print_solution(solution)
    print("Moves:", len(solution) - 1)
    print("States Expanded:", expanded)

    run_random_tests()
```

```python
import time
import random
import string
from collections import deque

class GraphProblem:
    def __init__(self, graph, start, goal):
        self.graph = graph
        self.start = start
        self.goal = goal

    def initial_state(self):
        return self.start

    def goal_test(self, state):
        return state == self.goal

    def actions(self, state):
        return self.graph.get(state, [])


class BreadthFirstSearch:
    def __init__(self, problem):
        self.problem = problem
        self.nodes_expanded = 0
        self.max_memory = 0

    def search(self):
        start_time = time.time()
        queue = deque([[self.problem.initial_state()]])
        visited = set([self.problem.initial_state()])

        while queue:
            self.max_memory = max(self.max_memory, len(queue))
            path = queue.popleft()
            node = path[-1]

            if self.problem.goal_test(node):
                end_time = time.time()
                return path, self.nodes_expanded, self.max_memory, end_time - start_time

            self.nodes_expanded += 1

            for neighbor in self.problem.actions(node):
                if neighbor not in visited:
                    visited.add(neighbor)
                    queue.append(path + [neighbor])

        end_time = time.time()
        return None, self.nodes_expanded, self.max_memory, end_time - start_time


class DepthFirstSearch:
    def __init__(self, problem):
        self.problem = problem
        self.nodes_expanded = 0
        self.max_memory = 0
        self.start_time = None

    def search(self):
        self.start_time = time.time()
        path = self._dfs(self.problem.initial_state(), [self.problem.initial_state()])
        end_time = time.time()
        return path, self.nodes_expanded, self.max_memory, end_time - self.start_time

    def _dfs(self, node, path):
        self.nodes_expanded += 1
        self.max_memory = max(self.max_memory, len(path))

        if self.problem.goal_test(node):
            return path

        for neighbor in self.problem.actions(node):
            if neighbor not in path:
                result = self._dfs(neighbor, path + [neighbor])
                if result is not None:
                    return result

        return None


def generate_random_graph(nodes, max_edges=3):
    graph = {}
    for node in nodes:
        neighbors = random.sample(nodes, random.randint(1, min(max_edges, len(nodes))))
        neighbors = [n for n in neighbors if n != node]
        graph[node] = neighbors
    return graph


def run_comparison_tests(test_cases=3):
    for i in range(test_cases):
        nodes = random.sample(string.ascii_uppercase, random.randint(6, 9))
        graph = generate_random_graph(nodes)
        start = random.choice(nodes)
        goal = random.choice(nodes)

        problem = GraphProblem(graph, start, goal)

        bfs = BreadthFirstSearch(problem)
        dfs = DepthFirstSearch(problem)

        bfs_path, bfs_expanded, bfs_memory, bfs_time = bfs.search()
        dfs_path, dfs_expanded, dfs_memory, dfs_time = dfs.search()

        print(f"\nTest Case {i + 1}")
        print("Graph:", graph)
        print("Start:", start, "Goal:", goal)
        print("\nAlgorithm | Path Length | Nodes Expanded | Max Memory | Time")
        print("BFS       |", 
              len(bfs_path) if bfs_path else None, "|",
              bfs_expanded, "|",
              bfs_memory, "|",
              round(bfs_time, 6))
        print("DFS       |", 
              len(dfs_path) if dfs_path else None, "|",
              dfs_expanded, "|",
              dfs_memory, "|",
              round(dfs_time, 6))


if __name__ == "__main__":
    graph = {
        'A': ['B', 'C'],
        'B': ['D', 'E'],
        'C': ['F'],
        'D': [],
        'E': ['F'],
        'F': ['G'],
        'G': []
    }

    problem = GraphProblem(graph, 'A', 'G')

    bfs = BreadthFirstSearch(problem)
    dfs = DepthFirstSearch(problem)

    bfs_result = bfs.search()
    dfs_result = dfs.search()

    print("Fixed Graph Comparison")
    print("Algorithm | Path Length | Nodes Expanded | Max Memory | Time")
    print("BFS       |",
          len(bfs_result[0]), "|",
          bfs_result[1], "|",
          bfs_result[2], "|",
          round(bfs_result[3], 6))
    print("DFS       |",
          len(dfs_result[0]), "|",
          dfs_result[1], "|",
          dfs_result[2], "|",
          round(dfs_result[3], 6))

    run_comparison_tests()
```
