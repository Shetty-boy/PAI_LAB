### Experiment 1: Cybersecurity Threat Detection using Logical Inference and Resolution

**Problem Statement:**

Modern cybersecurity systems often rely on rule-based reasoning to detect threats based on system activity.  
Unlike machine learning models, these systems provide clear and explainable decisions using logical inference.

Design a Cybersecurity Threat Detection Agent that determines whether a system intrusion has occurred using propositional logic and resolution.

The system takes:
* Facts (observed events)
* Rules (security knowledge)

The agent must:
* Convert rules into Conjunctive Normal Form (CNF)
* Apply resolution-based theorem proving
* Determine whether:
  **IntrusionDetected = TRUE or FALSE**

---

### Input Specification

**Facts (Percepts):**
Examples:
* MultipleFailedLogins
* LoginFromUnknownIP
* AccessToSensitiveFiles
* AdminPrivileges

**Rules (Knowledge Base):**
Examples:
* MultipleFailedLogins ∧ LoginFromUnknownIP → SuspiciousLogin  
* SuspiciousLogin ∧ AdminPrivileges → HighRiskIntrusion  
* HighRiskIntrusion → IntrusionDetected  
* AccessToSensitiveFiles ∧ ¬AdminPrivileges → IntrusionDetected  

---

### Task Requirements

1. Represent all rules using propositional logic.
2. Define:
    * Atomic propositions
    * Implications
    * Negations
3. Convert all rules into Conjunctive Normal Form (CNF).
4. Apply resolution-based inference to determine intrusion.

---

### Part A: Knowledge Representation

* Represent rules using propositional logic expressions.
* Identify:
    * Atomic propositions (e.g., MultipleFailedLogins)
    * Logical operators (∧, ∨, ¬, →)

---

### Part B: CNF Conversion

Convert all rules into CNF using the following steps:

1. Eliminate implications  
2. Move negations inward (De Morgan’s laws)  
3. Standardize expressions  
4. Convert into conjunction of disjunctions (CNF)

---

### Part C: Resolution Algorithm

1. Add negation of the query:
   
`¬IntrusionDetected`

2. Apply resolution rule:

`(A ∨ B), (¬B ∨ C) ⇒ (A ∨ C)`


3. Continue until:
* Empty clause is derived → TRUE (Intrusion detected)
* No new clauses → FALSE

---

### Program Requirements

**Input:**
* List of facts
* List of rules

**Output:**
* CNF clauses
* Resolution steps (intermediate clauses)
* Final conclusion:
  
`IntrusionDetected: TRUE / FALSE`


---

### Scenario Testing

Test the system with the following scenarios:

**Scenario 1: Low Risk**
* MultipleFailedLogins  
* Expected Output: No intrusion

**Scenario 2: Suspicious Activity**
* MultipleFailedLogins  
* LoginFromUnknownIP  
* Expected Output: Suspicious but no confirmed intrusion

**Scenario 3: High Risk Intrusion**
* MultipleFailedLogins  
* LoginFromUnknownIP  
* AdminPrivileges  
* Expected Output:

`IntrusionDetected = TRUE`


---

### Sample Output Format

**CNF Clauses:**
1. ¬MultipleFailedLogins ∨ ¬LoginFromUnknownIP ∨ SuspiciousLogin  
2. ¬SuspiciousLogin ∨ ¬AdminPrivileges ∨ HighRiskIntrusion  
3. ¬HighRiskIntrusion ∨ IntrusionDetected  

**Resolution Steps:**
* Derived: IntrusionDetected  

**Final Output:**



```python
from itertools import combinations

def resolve(ci, cj):
    resolvents = []
    for di in ci:
        for dj in cj:
            if di == ('¬' + dj) or ('¬' + di) == dj:
                new_clause = set(ci) | set(cj)
                new_clause.discard(di)
                new_clause.discard(dj)
                resolvents.append(tuple(new_clause))
    return resolvents


def resolution(kb, query):
    clauses = kb.copy()
    clauses.append(tuple(['¬' + query]))

    print("\nInitial Clauses:")
    for c in clauses:
        print(c)

    new = set()

    while True:
        pairs = list(combinations(clauses, 2))

        for (ci, cj) in pairs:
            resolvents = resolve(ci, cj)

            for r in resolvents:
                if len(r) == 0:
                    print("\nDerived empty clause → IntrusionDetected = TRUE")
                    return True
                new.add(r)

        if new.issubset(set(clauses)):
            print("\nNo new clauses → IntrusionDetected = FALSE")
            return False

        for c in new:
            if c not in clauses:
                print("Derived:", c)
                clauses.append(c)
```

```python
# CNF Clauses
kb = [
    ('¬MFL', '¬LUI', 'SL'),
    ('¬SL', '¬AP', 'HRI'),
    ('¬HRI', 'ID'),
    ('¬ASF', 'AP', 'ID')
]

# Facts 
facts = [
    ('MFL',),
    ('LUI',),
    ('AP',)
]

# Add facts to KB
kb.extend(facts)

# Run resolution
result = resolution(kb, 'ID')

print("\nFinal Result:")
print("IntrusionDetected =", result)
```

### 2. Wastewater Treatment Decision using Logical Inference

Design a logical agent to determine TreatmentRequired = TRUE / FALSE using:
* Facts (sensor readings)
* Rules (domain knowledge)
* CNF conversion
* Resolution

### Input

**Facts:**
* HighBOD
* LowDO
* HighTurbidity
* ToxicChemicals
* pHImbalance

**Rules:**
* HighBOD ∧ LowDO → OrganicPollution
* HighTurbidity ∨ ToxicChemicals → Contamination
* OrganicPollution ∧ Contamination → SeverePollution
* SeverePollution → TreatmentRequired
* pHImbalance → TreatmentRequired

### Tasks

1. Represent rules in propositional logic  
2. Convert all rules into CNF  
3. Apply resolution to check if TreatmentRequired is TRUE  

### Test Cases

**Scenario 1:** HighBOD, Output: FALSE  

**Scenario 2:** HighBOD, LowDO, HighTurbidity  

**Scenario 3:** ToxicChemicals  
Output: TRUE  

### Output

CNF clauses, Resolution steps  

Final result:

TreatmentRequired = TRUE / FALSE

```python
from itertools import combinations

def resolve(ci, cj):
    resolvents = []
    for di in ci:
        for dj in cj:
            if di == ('¬' + dj) or ('¬' + di) == dj:
                new_clause = set(ci) | set(cj)
                new_clause.discard(di)
                new_clause.discard(dj)
                resolvents.append(tuple(new_clause))
    return resolvents


def resolution(kb, query):
    clauses = kb.copy()
    clauses.append(('¬' + query,))

    print("\nInitial Clauses:")
    for c in clauses:
        print(c)

    new = set()

    while True:
        pairs = list(combinations(clauses, 2))

        for (ci, cj) in pairs:
            resolvents = resolve(ci, cj)

            for r in resolvents:
                if len(r) == 0:
                    print("\nDerived empty clause → TreatmentRequired = TRUE")
                    return True
                if r not in clauses:
                    print("Derived:", r)
                    new.add(r)

        if new.issubset(set(clauses)):
            print("\nNo new clauses → TreatmentRequired = FALSE")
            return False

        for c in new:
            if c not in clauses:
                clauses.append(c)
```

```python
# CNF Clauses
kb = [
    ('¬HB', '¬LD', 'OP'),
    ('¬HT', 'C'),
    ('¬TC', 'C'),
    ('¬OP', '¬C', 'SP'),
    ('¬SP', 'TR'),
    ('¬PH', 'TR'),

    # Optional rule for Scenario 3
    ('¬TC', 'TR')
]

def run_test(facts, name):
    print("\n", name)

    test_kb = kb.copy()
    for f in facts:
        test_kb.append((f,))

    result = resolution(test_kb, 'TR')
    print("Final Result:", result)


# Scenario 1
run_test(['HB'], "Scenario 1")

# Scenario 2
run_test(['HB', 'LD', 'HT'], "Scenario 2")

# Scenario 3
run_test(['TC'], "Scenario 3")
```
