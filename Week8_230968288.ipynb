```python
class SmartLightingAgent:
    def __init__(self):
        self.knowledge_base = {
            'MotionDetected': False,
            'IsDark': False,
            'ManualOverride': False
        }

    def update_knowledge(self, motion_detected, is_dark, manual_override):
        self.knowledge_base['MotionDetected'] = motion_detected
        self.knowledge_base['IsDark'] = is_dark
        self.knowledge_base['ManualOverride'] = manual_override

    def infer_action(self):
        if self.knowledge_base['ManualOverride']:
            return 'Lights OFF (Manual Override)'
        
        if self.knowledge_base['IsDark'] and self.knowledge_base['MotionDetected']:
            return 'Lights ON'
        
        if not self.knowledge_base['MotionDetected']:
            return 'Lights OFF'
        
        return 'No Action'

# Example usage:
lighting_agent = SmartLightingAgent()

# Set sensor readings
lighting_agent.update_knowledge(motion_detected=True, is_dark=True, manual_override=False)
print(lighting_agent.infer_action())  # Should print: 'Lights ON'

lighting_agent.update_knowledge(motion_detected=False, is_dark=True, manual_override=False)
print(lighting_agent.infer_action())  # Should print: 'Lights OFF'

lighting_agent.update_knowledge(motion_detected=False, is_dark=False, manual_override=True)
print(lighting_agent.infer_action())  # Should print: 'Lights OFF (Manual Override)'
```

```python
class MedicalDiagnosisAgent:
    def __init__(self):
        self.knowledge_base = {
            'Fever': False,
            'Cough': False,
            'ChestPain': False,
            'Flu': False,
            'Pneumonia': False,
            'HospitalizationRequired': False
        }

    def update_knowledge(self, fever, cough, chest_pain):
        self.knowledge_base['Fever'] = fever
        self.knowledge_base['Cough'] = cough
        self.knowledge_base['ChestPain'] = chest_pain

    def forward_chain(self):
        if self.knowledge_base['Fever'] and self.knowledge_base['Cough']:
            self.knowledge_base['Flu'] = True
        if self.knowledge_base['Fever'] and self.knowledge_base['Cough'] and self.knowledge_base['ChestPain']:
            self.knowledge_base['Pneumonia'] = True
        if self.knowledge_base['Pneumonia']:
            self.knowledge_base['HospitalizationRequired'] = True

    def infer_diseases(self):
        diseases = []
        if self.knowledge_base['Flu']:
            diseases.append('Flu')
        if self.knowledge_base['Pneumonia']:
            diseases.append('Pneumonia')
        return diseases

    def infer_actions(self):
        actions = []
        if self.knowledge_base['HospitalizationRequired']:
            actions.append('Hospitalization Required')
        return actions

medical_agent = MedicalDiagnosisAgent()

medical_agent.update_knowledge(fever=True, cough=True, chest_pain=True)

medical_agent.forward_chain()

diseases = medical_agent.infer_diseases()
print(f"Possible Diseases: {diseases}")

actions = medical_agent.infer_actions()
print(f"Required Actions: {actions}")
```

```python
class TrafficViolationAgent:
    def __init__(self):
        self.knowledge_base = {
            'SpeedAboveLimit': False,
            'SignalRed': False,
            'VehicleCrossedLine': False,
            'EmergencyVehicle': False,
            'SpeedViolation': False,
            'SignalViolation': False
        }

    def update_knowledge(self, speed_above_limit, signal_red, vehicle_crossed_line, emergency_vehicle):
        self.knowledge_base['SpeedAboveLimit'] = speed_above_limit
        self.knowledge_base['SignalRed'] = signal_red
        self.knowledge_base['VehicleCrossedLine'] = vehicle_crossed_line
        self.knowledge_base['EmergencyVehicle'] = emergency_vehicle

    def infer_violations(self):
        if self.knowledge_base['EmergencyVehicle']:
            self.knowledge_base['SpeedViolation'] = False
            self.knowledge_base['SignalViolation'] = False
        else:
            if self.knowledge_base['SpeedAboveLimit']:
                self.knowledge_base['SpeedViolation'] = True
            else:
                self.knowledge_base['SpeedViolation'] = False

            if self.knowledge_base['SignalRed'] and self.knowledge_base['VehicleCrossedLine']:
                self.knowledge_base['SignalViolation'] = True
            else:
                self.knowledge_base['SignalViolation'] = False

    def decide_enforcement_action(self):
        violations = []
        if self.knowledge_base['SpeedViolation']:
            violations.append('Speed Violation')
        if self.knowledge_base['SignalViolation']:
            violations.append('Signal Violation')
        if violations:
            return f"Issuing Ticket for: {', '.join(violations)}"
        return 'No Violation Detected'

traffic_agent = TrafficViolationAgent()

traffic_agent.update_knowledge(speed_above_limit=True, signal_red=True, vehicle_crossed_line=True, emergency_vehicle=False)
traffic_agent.infer_violations()
print(traffic_agent.decide_enforcement_action())  

traffic_agent.update_knowledge(speed_above_limit=False, signal_red=True, vehicle_crossed_line=True, emergency_vehicle=True)
traffic_agent.infer_violations()
print(traffic_agent.decide_enforcement_action())  
```

```python
class CybersecurityIntrusionAgent:
    def __init__(self):
        self.knowledge_base = {
            'MultipleFailedLogins': False,
            'LoginFromUnknownIP': False,
            'AccessToSensitiveFiles': False,
            'AdminPrivileges': False,
            'SuspiciousLogin': False,
            'HighRiskIntrusion': False,
            'TriggerAlert': False
        }

    def update_knowledge(self, multiple_failed_logins, login_from_unknown_ip, access_to_sensitive_files, admin_privileges):
        self.knowledge_base['MultipleFailedLogins'] = multiple_failed_logins
        self.knowledge_base['LoginFromUnknownIP'] = login_from_unknown_ip
        self.knowledge_base['AccessToSensitiveFiles'] = access_to_sensitive_files
        self.knowledge_base['AdminPrivileges'] = admin_privileges

    def forward_chain(self):
        if self.knowledge_base['MultipleFailedLogins'] and self.knowledge_base['LoginFromUnknownIP']:
            self.knowledge_base['SuspiciousLogin'] = True
        if self.knowledge_base['SuspiciousLogin'] and self.knowledge_base['AdminPrivileges']:
            self.knowledge_base['HighRiskIntrusion'] = True
        if self.knowledge_base['HighRiskIntrusion']:
            self.knowledge_base['TriggerAlert'] = True

    def infer_threat_level(self):
        if self.knowledge_base['HighRiskIntrusion']:
            return 'High Risk Intrusion'
        return 'No Threat Detected'

    def decide_defensive_action(self):
        if self.knowledge_base['TriggerAlert']:
            return 'Trigger Alert'
        return 'No Action'

# Example usage:
cybersecurity_agent = CybersecurityIntrusionAgent()

cybersecurity_agent.update_knowledge(multiple_failed_logins=True, login_from_unknown_ip=True, access_to_sensitive_files=False, admin_privileges=True)
cybersecurity_agent.forward_chain()
print(cybersecurity_agent.infer_threat_level())  # Should print: High Risk Intrusion
print(cybersecurity_agent.decide_defensive_action())  # Should print: Trigger Alert

cybersecurity_agent.update_knowledge(multiple_failed_logins=False, login_from_unknown_ip=False, access_to_sensitive_files=False, admin_privileges=False)
cybersecurity_agent.forward_chain()
print(cybersecurity_agent.infer_threat_level())  # Should print: No Threat Detected
print(cybersecurity_agent.decide_defensive_action())  # Should print: No Action
```

```python
class WarehouseRobotAgent:
    def __init__(self):
        self.knowledge_base = {
            'ObstacleAhead': False,
            'PathClear': False,
            'CarryingFragileItem': False,
            'HumanNearby': False,
            'DestinationReached': False
        }

    def update_knowledge(self, obstacle_ahead, path_clear, carrying_fragile_item, human_nearby, destination_reached):
        self.knowledge_base['ObstacleAhead'] = obstacle_ahead
        self.knowledge_base['PathClear'] = path_clear
        self.knowledge_base['CarryingFragileItem'] = carrying_fragile_item
        self.knowledge_base['HumanNearby'] = human_nearby
        self.knowledge_base['DestinationReached'] = destination_reached

    def infer_action(self):
        if self.knowledge_base['ObstacleAhead']:
            return 'Stop'
        if self.knowledge_base['CarryingFragileItem'] and self.knowledge_base['ObstacleAhead']:
            return 'Slow Down'
        if self.knowledge_base['HumanNearby']:
            return 'Reduce Speed'
        if self.knowledge_base['PathClear'] and not self.knowledge_base['HumanNearby']:
            return 'Move Forward'
        return 'No Action'

robot_agent = WarehouseRobotAgent()

robot_agent.update_knowledge(obstacle_ahead=True, path_clear=False, carrying_fragile_item=True, human_nearby=False, destination_reached=False)
print(robot_agent.infer_action())  # Should print: Stop

robot_agent.update_knowledge(obstacle_ahead=False, path_clear=True, carrying_fragile_item=True, human_nearby=False, destination_reached=False)
print(robot_agent.infer_action())  # Should print: Slow Down

robot_agent.update_knowledge(obstacle_ahead=False, path_clear=True, carrying_fragile_item=False, human_nearby=True, destination_reached=False)
print(robot_agent.infer_action())  # Should print: Reduce Speed

robot_agent.update_knowledge(obstacle_ahead=False, path_clear=True, carrying_fragile_item=False, human_nearby=False, destination_reached=False)
print(robot_agent.infer_action())  # Should print: Move Forward
```
